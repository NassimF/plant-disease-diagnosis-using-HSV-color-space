# Segmentation with color spaces HSV & RGB

## import libraries 

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from matplotlib import colors
import os
```

## Zoom function

```python
def zoom(frame):
    scale = 20
    #get the webcam size
    height, width, channels = frame.shape

    #prepare the crop
    centerX,centerY=int(height/2),int(width/2)
    radiusX,radiusY= int(scale*height/100),int(scale*width/100)

    minX,maxX=centerX-radiusX,centerX+radiusX
    minY,maxY=centerY-radiusY,centerY+radiusY

    cropped = frame[minX:maxX, minY:maxY]
    resized_cropped = cv2.resize(cropped, (width, height)) 
    
    return resized_cropped
    
```

## Extract, zoom and show frames

```python

cap = cv2.VideoCapture(r"C:\Users\asus\Desktop\notebooks\image-segmentation\vids\3.mp4")

images = []



while(cap.isOpened()):
    ret, frame = cap.read()
    
    if ret == False:
        print('Error! Could not read frame')
        break

    #zoom and show images before color conversion
    frame = zoom(frame)
    cv2.imshow('before', frame) 

    # Exit if "q" is pressed
    if cv2.waitKey(25) & 0xFF == ord('q'):
      break

    #convert color and add image to a list
    image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    images.append(image)
    
    
    


    
cap.release()
cv2.destroyAllWindows()



```

## Function to create and apply masks

```python
def segment_image(image):
    ''' Attempts to segment the yellow spots out of the provided image '''

    # Convert the image into HSV
    hsv_image = cv2.cvtColor(image, cv2.COLOR_RGB2HSV)

    
    # Set a yellow range
    light_yellow = (47/2, (38/100) * 255, (60/100) * 255) 
    dark_yellow = (67/2, (100/100) * 255, (100/100) * 255)
    

    # Apply the  mask
    mask = cv2.inRange(hsv_image, light_yellow, dark_yellow)

    
    result = cv2.bitwise_and(image, image, mask=mask)
    return result
```

## Get the results

```python
results = [segment_image(image) for image in images]
```

# Plot and show images

```python
#plot and save result images
for i in range(len(images)):
    print('i is=',i)
    plt.subplot(1, 2, 1)
    plt.imshow(images[i])
    plt.xlabel('original')
    plt.subplot(1, 2, 2)
    plt.imshow(results[i])
    plt.xlabel('masked')
    plt.savefig(r'C:\Users\asus\Desktop\notebooks\image-segmentation\result vid\res_vid3_'+ str(i+1))
    plt.show()

# #show video
for res in results:
    res = cv2.cvtColor(res, cv2.COLOR_HSV2BGR)
    cv2.imshow('After', res) 
    
    if cv2.waitKey(25) & 0xFF == ord('q'):
      break


cap.release()
cv2.destroyAllWindows()
  



```
